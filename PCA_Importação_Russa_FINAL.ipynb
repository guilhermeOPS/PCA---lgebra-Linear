```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('/content/russia-import.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# Calcular a Matriz de Covariância
d1 = dataset.drop(columns=['ANO'])
d2 = d1[:-2]
dataset_atualizado = d2.select_dtypes(include=[float, int])


# Padronizar os dados
padronizar = StandardScaler()
dados_normalizados = padronizar.fit_transform(dataset_atualizado)

MCov = np.cov(dados_normalizados)
```

```python
d1
```

```python
dados_normalizados.T
```

```python
MCov
```

```python
# Encontrar autovalores e autovetores
autovalores, autovetores = np.linalg.eig(MCov)

print("\nAutovalores:")
print(autovalores)

print("\nAutovetores:")
print(autovetores)
```

```python
# Recolher os maiores autovalores e autovetores
maior_autova = np.argsort(autovalores)[::-1]

autova_decres = autovalores[maior_autova]
autove_decres = autovetores[:, maior_autova]
```

```python
autove_decres
```

```python
# Novo dataset a partir dos autovetores selecionados
componentes = 2

busca_pelo_autovetor = autove_decres[:, :componentes]
busca_pelo_autovetor

# Projetar os dados no novo espaço (multiplicar os dados normalizados pelos autovetores)
novo_dataset = np.matmul(dados_normalizados.T, busca_pelo_autovetor)
```

```python
novo_dataset
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.scatter(novo_dataset[:, 0], novo_dataset[:, 1], color='blue')
plt.title("Projeção dos Dados sobre a Importação Russa")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.grid(True)
plt.show()
```
